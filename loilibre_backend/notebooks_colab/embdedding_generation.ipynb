```python
## In this code we will compute the embedding for each token
from sentence_transformers import SentenceTransformer, util
```

```python
model = SentenceTransformer('multi-qa-mpnet-base-dot-v1')

```

```python
ds_name = "Forbu14/LoiLibre"

from datasets import load_dataset
ds = load_dataset("Forbu14/LoiLibre", split="train")
```

```python
batch_size = 16

nb_batch = len(ds) // batch_size
embedding_list = []
embedding_index = []

from tqdm import tqdm
import numpy as np

for idx_batch in tqdm(range(nb_batch)):
  batch = ds[(idx_batch * batch_size):((idx_batch + 1) * batch_size)]
  texts = batch["text"]
  embeddings = model.encode(texts)

  embedding_list.append(embeddings)
  embedding_index = embedding_index + list(range(idx_batch * batch_size, (idx_batch + 1) * batch_size))

embedding_array = np.concatenate(embedding_list)
embedding_df = list(embedding_array)
```

```python
import pandas as pd
df_embedding = pd.DataFrame({"embedding": embedding_df, "index" : embedding_index})
df_embedding.to_parquet("embedding.parquet")
```
